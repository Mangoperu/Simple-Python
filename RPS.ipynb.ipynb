```python
import random
```

```python
def computer():
    possible = ["rock" , "paper" , "scissor"]
    comp = random.choice(possible)
    return comp
```

```python
man = input('Choose one out of rock, paper , scissor')
comp = computer()
print(comp)

```

```python
def result():
    if man == "rock":
        if comp == "rock":
            print("draw")
        elif comp == "paper":  
            print("lost")
        else:
            print("won")
            
    
    elif man == "paper":
        if comp == "rock":
            print("won")
        elif comp == "paper":
            print("draw")
        else:
            print("lost")
            
    
    else: 
        if comp == "rock":
            print("lost")
        elif comp == "paper":
            print("won")
        else:
            print("draw")
            
    
        
```
